O código abaixo tem como objetivo construir uma senóide da forma "y(t) = A*sen(omega*t)". Nessa representação, "A" representa a amplitude da onda e "omega" representa sua velocidade angular, dada pela frequência da onda multiplicada por "2*pi", ou seja, "omega = 2*pi*freq". Para essa construção, a velocidade angular foi estabelecida atribuindo um valor de 1000Hz para frequência da onda "freq", isto é, a onda percorre 1000 ciclos por unidade de tempo. Para a variável de tempo "t", que estabelece a quantidade de amostras da onda em questão, foi construído um vetor com "N" elementos, onde "N" é igual a 10% do valor da taxa de amostragem "Fs". Logo, o vetor "t_vec" possui 800 elementos, considerando uma taxa de amostragem "Fs" igual a 8000 amostras por unidade de tempo. Cada elemento é normalizado por um fator "T", dado pelo inverso da taxa de amostragem "Fs", ou seja, "T = 1/8000". Isso significa que a onda será construída a partir de uma taxa de amostragem de 8000 amostras por unidade de tempo. Como o intervalo de tempo considerado é somente 10% de uma unidade de tempo, serão captadas 800 amostras do sinal, a mesma quantidade de elementos do vetor de unidades de tempo "t_vec". Cada elemento desse vetor é multiplicado pelo período de amostragem, dado por 1/8000, pois as amostras devem estar limitadas entre os valores 0/8000 = 0 e 800/8000 = 0.1, dado o intervalo de tempo em questão. Como a frequência dessa onda é de 1000Hz e o intervalo de tempo é de 10% de uma unidade de tempo, então o gráfico mostra uma senóide com 1000/10 = 100 ciclos completos. Além da quantidade de ciclos igual a 100, temos também outros atributos definidos, como a amplitude igual a 1 e a fase igual a 0.

```
from IPython.display import Audio
import numpy as np
import matplotlib.pyplot as plt

Fs = 8000
T = 1 / Fs
t = 0.1
N = Fs * t
freq = 1000
omega = 2 * np.pi * freq
t_vec = np.arange(N) * T
y = np.sin(omega * t_vec)
plt.plot(t_vec, y)
plt.show()
```

A taxa de amostragem "Fs" estabelece a quantidade de amostras que serão captadas em um dado intervalo de tempo. Portanto, esse parâmetro diz respeito a fidelidadde, ou a qualidade, da representação do áudio. Nesse exemplo, a qualidade da representação gráfica do áudio. Para os três exemplos anteriores, foi observado, com base nos gráficos, que o aumento da taxa de amostragem "Fs", nesse caso, não causou muita variação na representação do sinal. Isso significa que, para essa onda, considerando todos os seus parâmetros, não é necessária uma taxa de amostragem tão alta para uma representação fiel do áudio. A frequência de amostragem deve ser maior que o dobro da frequência do sinal (Teorema de Nyquist), porém, não é necessário que essa frequência seja muito maior que o dobro.

```
#c.1) Alterando a amplitude do sinal para 10 vezes superior à amplitude inicial 
#Alterando a amplitude da onda dessa forma, a mesma varia entre -10 e 10.
from IPython.display import Audio
import numpy as np
import matplotlib.pyplot as plt

A = 10
Fs = 8000
T = 1 / Fs
t = 0.1
N = Fs * t
freq = 1000
omega = 2 * np.pi * freq
t_vec = np.arange(N) * T
y = A * np.sin(omega * t_vec)
plt.plot(t_vec, y)
plt.show()
```
