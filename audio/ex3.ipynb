O código abaixo tem como objetivo normalizar um arquivo de áudio, limitando sua amplitude entre os valores -1 e 1. Primeiramente, a amostra do sinal é submetida a uma troca de tipo, passando de "int" para "float". Assim, as operações matemáticas necessárias serão possíveis. Após, todos os elementos da amostra são divididos pelo módulo do valor máximo. Dessa forma, o valor mais alto encontrado na amostra será menor ou igual a 1 e o menor valor será maior ou igual a -1. Finalmente, a amostra é subtraída do valor médio, finalizando o processo de normalização. Esse procedimento altera somente a amplitude do sinal. Os outros atributos, como frequência, fase e taxa de amostragem, permanecem inalterados. De forma resumida, a forma da onda não muda, somente sua amplitude.

```
#Análise do áudio "male.wav" antes da aplicação das etapas de normalização
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile
from scipy.io.wavfile import read, write
from google.colab import drive
from IPython.display import Audio

drive.mount('/content/gdrive')

root_path = '/content/gdrive/My Drive/Colab Notebooks/'
sr_male, s_male = read(root_path + 'dados/male.wav')

plt.figure()
plt.plot(s_male)
plt.xlabel('Amostra')
plt.ylabel('Amplitude')
plt.title('Forma de onda male.wav - Não normalizado')
plt.show()

Audio(s_male, rate = sr_male)
```

```
#Análise do áudio "male.wav" após a aplicação das etapas de normalização
b_normalize = True

if b_normalize:
  s_male = s_male.astype(np.float32)
  s_male = (s_male / np.max(np.abs(s_male)))
  s_male -= np.mean(s_male)

plt.figure()
plt.plot(s_male)
plt.xlabel('Amostra')
plt.ylabel('Amplitude')
plt.title('Forma de onda male.wav - Normalizado')
plt.show()

Audio(s_male, rate = sr_male)
```
