O código abaixo tem como objetivo criar um áudio, baseado em alguns fatores, tais como:

A amplitude ("amplitude") e a frequência da onda ("fs");

Os valores de instantes de tempo em que as amostras da onda foram capturadas ("t"), que vão de 0 até 1, espaçados por um valor de 1/44100;

E, finalmente, a taxa de amostragem ("samplerate").

```
#Abaixo, o código com parâmetros originais (samplerate = 44100, fs = 100).
from scipy.io.wavfile import write
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
from IPython.display import Audio
%matplotlib inline

drive.mount('/content/gdrive')

samplerate = 44100
fs = 100
t = np.linspace(0., 1., samplerate)
amplitude = np.iinfo(np.int16).max
data = amplitude * np.sin(2. * np.pi * fs * t)

root_path = '/content/gdrive/My Drive/Colab Notebooks/'
write(root_path + "dados/teste1.wav", samplerate, data)

plt.figure()
plt.plot(t,data)
plt.xlabel('Amostra')
plt.ylabel('Amplitude')
plt.title('Forma de onda teste1.wav')
plt.show()

Audio(data, rate = samplerate)
```

Portanto, aumentar muito essa taxa pode não ser interessante, visto que o áudio não sofre grandes mudanças e mais dados estão sendo captados desnecessariamente, tornando o armazenamento e o processamento mais caros.
