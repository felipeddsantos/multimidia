```
#Áudio de tom mais alto, mono, 8kHz de taxa de amostragem: análise forma de onda mais detalhada
from google.colab import drive
from scipy.io.wavfile import read, write
from IPython.display import Audio
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
%matplotlib inline

root_path = '/content/gdrive/My Drive/Colab Notebooks/'
Fs_alto, data_alto = read(root_path + 'dados/tom_alto.wav')

data_alto = data_alto[6000:6200]

plt.figure()
plt.plot(data_alto)
plt.xlabel('Amostra')
plt.ylabel('Amplitude')
plt.title('Forma de onda tom_alto.wav')
plt.show()
```

```
#Áudio de tom mais baixo, mono, 8kHz de taxa de amostragem: análise da forma de onda mais detalhada
root_path = '/content/gdrive/My Drive/Colab Notebooks/'
Fs_baixo, data_baixo = read(root_path + 'dados/tom_baixo.wav')

data_baixo = data_baixo[6000:6200]

plt.figure()
plt.plot(data_baixo)
plt.xlabel('Amostra')
plt.ylabel('Amplitude')
plt.title('Forma de onda tom_baixo.wav')
plt.show()
```

Como é observado, em um mesmo intervalo de tempo, a forma de onda do tom mais baixo tem menor frequência, enquanto que a forma de onda do tom mais alto tem maior frequência. Esse resultado é esperado, considerando que um tom mais baixo é caracterizado por uma baixa frequência e um tom mais alto é caracterizado por uma frequência mais alta.

```
#Áudio de tom mais alto, mono, 8kHz de taxa de amostragem: análise do espectograma 
from google.colab import drive
from scipy.io.wavfile import read, write
from IPython.display import Audio
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
%matplotlib inline

root_path = '/content/gdrive/My Drive/Colab Notebooks/'
Fs_alto, data_alto = read(root_path + 'dados/tom_alto.wav')

frequencies, times, spectrogram = signal.spectrogram(data_alto, Fs_alto)
plt.pcolormesh(times, frequencies, spectrogram)
plt.imshow(spectrogram, origin = 'lower', aspect = 'auto')
plt.ylabel('Frequência [Hz]')
plt.xlabel('Tempo [sec]')
plt.title('Espectograma tom_alto.wav')
plt.show()
```

```
#Áudio de tom mais baixo, mono, 8kHz de taxa de amostragem: análise do espectograma 
Fs_baixo, data_baixo = read(root_path + 'dados/tom_baixo.wav')

frequencies, times, spectrogram = signal.spectrogram(data_baixo, Fs_baixo)
plt.pcolormesh(times, frequencies, spectrogram)
plt.imshow(spectrogram, origin = 'lower', aspect = 'auto')
plt.ylabel('Frequência [Hz]')
plt.xlabel('Tempo [sec]')
plt.title('Espectograma tom_baixo.wav')
plt.show()
```

Como pode ser observado, o espectograma do áudio de tom mais alto possui faixas em frequências mais altas. As faixas mais altas são mais acentuadas e algumas delas chegam a valores próximos de 40. Por outro lado, o espectograma do áudio de tom mais baixo possui faixas em frequências mais baixas. As faixas mais baixas são mais acentuadas e as maiores faixas de frequência não ultrapassam 20. Portanto, comparando os espectogramas, o áudio de tom mais alto possui faixas de frequências mais altas (a maior chega perto de 40), enquanto que o áudio de tom mais baixo possui faixas de frequências mais baixas (a maior não passa de 20). Como explanado anteriormente, esse resultado é esperado e condiz com as formas de ondas dos áudios, considerando que o tons mais baixos possuem menores frequências e tons mais altos possuem maiores frequências.
