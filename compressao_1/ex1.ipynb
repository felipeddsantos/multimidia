```
#Função que realiza o cálculo da entropia de um conjunto de dados armazenados em um vetor "V"
def entropy(V):

  #Obtenção da frequência de ocorrência de cada símbolo do alfabeto
  _, freq = np.unique(V, return_counts = True)

  #Obtenção do vetor de probabilidades, normalizando o vetor de frequências
  p = freq / len(V)

  #Variável para armazenamento da entropia
  etp = 0.0

  #Laço que analisa cada dado do vetor de probabilidades
  for i in range(len(p)):

    etp = etp - p[i] * np.log2(p[i])
  
  return etp
```

Utilizando a função definida acima, serão selecionados 3 arquivos de texto e 3 imagens em níveis de cinza para o cálculo da entropia.

```
#Declaração das bibliotecas necessárias, definição do diretório raíz e montagem do diretório
from google.colab import drive
import numpy as np
from PIL import Image
from numpy import asarray
import matplotlib.pyplot as plt

drive.mount('/content/gdrive')

root_path = '/content/gdrive/My Drive/Colab Notebooks/dados/'
```

```
#Imagem I

#Carregamento da imagem
im = Image.open(root_path + 'lena.tiff')

#Mostrar a imagem
plt.imshow(im, 'gray', origin = 'upper')
plt.show()

#Transformação da imagem em matriz
M_pixel = asarray(im)

#Transformação da matriz em vetor
V_pixel = np.matrix.flatten(M_pixel)

#Cálculo da entropia
print("Entropia da imagem:", entropy(V_pixel))
```
