```
#Função para ordenação de um vetor de símbolos "alf" em ordem decrescente de acordo com suas frequências "freq"
def ord(alf, freq):

  #Listas que serão armazenados os resultados referentes a "V" em ordem decrescente
  alf_ord = []
  freq_ord = []

  for i in range(len(alf)):

    #Índice do elemento de máxima frequência
    max = np.argmax(freq)

    #Obtenção do elemento do alfabeto de maior frequência para ser inserido no vetor de alfabeto ordenado
    alf_ord.append(alf[max])
    
    #Obtenção da maior frequência para ser inserido no vetor de frequência ordenada
    freq_ord.append(freq[max])

    #Eliminação do elemento já analisado
    freq[max] = -1

  return alf_ord, freq_ord
```

```
#Função para o cálculo da tabela de códigos de acordo com o algoritmo de Huffman. 
#O vetor "freq" contém as frequências dos símbolos do alfabeto. A cada iteração do algoritmo, os dois menores elementos são somados, de forma que passam a ocupar a última posição do vetor. 
#O vetor "alf_var" inicialmente contém o alfabeto da string a ser codificada. Com a junção das menores frequências do vetor "freq", ocorre também a junção dos símbolos de menores frequências do vetor "alf_var". 
#Logo, a cada passo do algoritmo, o vetor "alf_var" é modificado, de acordo com as modificações do vetor "freq".
#Os vetores "freq" e "alf_var" são sempre sincronizados, de forma que o i-ésimo símbolo do vetor "alf_var" tem uma frequência dada pelo i-ésimo elemento do vetor "freq".
#O vetor "alf_abs" possui o alfabeto da string. Esse vetor nunca é modificado e é utilizado para referência do vetor "cod".
#O vetor "cod" armazena os códigos resultantes do processo. 
#Os vetores "cod" e "alf_abs" são sincronizados, de modo que o i-ésimo elemento de "cod" armazena o código referente ao i-ésimo símbolo de "alf_abs". 
def huffman(alf_abs, alf_var, freq, cod):

  #Ordenação do alfabeto corrente e do vetor de frequências em ordem decrescente
  alf_var, freq = ord(alf_var, freq)

  #Obtenção das dimensões dos vetores ordenados
  n_simb = len(alf_var)
  n_freq = len(freq)

  #Laço para análise do elemento de menor frequência do vetor do alfabeto corrente
  for i in range(len(alf_var[n_simb - 1])):
    
    #Obtenção do elemento corrente, presente no vetor "alf_var"
    elem = alf_var[n_simb - 1][i]

    #Obtenção da posição do elemento corrente no vetor "alf_abs"
    pos = alf_abs.index(elem)

    #Incrementação do código do elemento corrente, presente no vetor "cod", na posição espeficicada anteriormente 
    cod[pos] = '1' + cod[pos]
  
  #Laço para análise do segundo elemento de menor frequência do vetor do alfabeto corrente
  for i in range(len(alf_var[n_simb - 2])):
    
    #Obtenção do elemento corrente, presente no vetor "alf_var"
    elem = alf_var[n_simb - 2][i]

    #Obtenção da posição do elemento corrente no vetor "alf_abs"
    pos = alf_abs.index(elem)

    #Incrementação do código do elemento corrente, presente no vetor "cod", na posição espeficicada anteriormente 
    cod[pos] = '0' + cod[pos]

  #Se o vetor "alf_var" possuir mais de dois elementos, o algoritmo faz modificações em "alf_var" e "freq" e prossegue na execução do algoritmo
  if n_simb > 2:
  
    #Junção das duas menores frequências e atribuição na penúltima posição de "freq"
    freq[n_freq - 2] = freq[n_freq - 2] + freq[n_freq - 1]

    #Claramente, como atribuímos as duas útimas frequências na penúltima posição, devemos deletar a última posição
    freq = np.delete(freq, n_freq - 1)

    #Assim como no vetor "freq", devemos unir os dois últimos caracteres do vetor "alf_var", formando um novo símbolo para o alfabeto modificado
    alf_var[n_simb - 2] = alf_var[n_simb - 2] + alf_var[n_simb - 1]

    #E, novamente, deletar a última posição
    alf_var = np.delete(alf_var, n_simb - 1)

    #Recursão do algoritmo com os vetores "alf_var" e "freq" modificados e "alf_abs" e "cod" não modificados
    huffman(alf_abs, alf_var, freq, cod)
```

```
#Função para codificação de um conjunto de dados "string", dado sua lista de códigos "cod" e seu alfabeto "alf"
def code(string, alf, cod):

  #String para armazenamento do resultado
  cod_string = ""

  for i in range(len(string)):

    #Obtenção do elemento corrente da string
    elem = string[i]
    
    #Obtenção da posição do elemento corrente no alfabeto
    pos = alf.index(elem)

    #Obtenção do código correspondente ao elemento corrente e junção do mesmo à codificação final
    cod_string += cod[pos]

  return cod_string
```

```
#Função que realiza o cálculo da entropia de um conjunto de dados armazenados em um vetor "V"
def entropy(V):

  #Obtenção da frequência de ocorrência de cada símbolo do alfabeto
  _, freq = np.unique(V, return_counts = True)

  #Obtenção do vetor de probabilidades, normalizando o vetor de frequências
  p = freq / len(V)

  #Variável para armazenamento da entropia
  etp = 0.0

  #Laço que analisa cada dado do vetor de probabilidades
  for i in range(len(p)):

    etp = etp - p[i] * np.log2(p[i])
  
  return etp
```

Utilizando o algoritmo de Huffman e as funções auxiliares definidas acima, 3 arquivos de texto distintos serão codificados. Para medir a eficiência dessa técnica, será calculada a taxa de compressão dos mesmos e será realizada uma comparação da quantidade média de bits do resultado com a entropia da informação original.

```
#Declaração das bibliotecas necessárias, definição do diretório raíz e montagem do diretório
from google.colab import drive
import numpy as np

drive.mount('/content/gdrive')

root_path = '/content/gdrive/My Drive/Colab Notebooks/dados/'
```

```
#Texto I

#Carregamento do texto
txt = open(root_path + 'brandenburg.txt', 'r')

#Obtenção do conteúdo do arquivo
string = txt.read()
txt.close()

print("Texto original:", string)

#Obtenção do tamanho da string
n_string = len(string)

#Conversão da string em lista
V_txt = list(string)

#Obtenção do alfabeto e das frequências da string
alf, freq = np.unique(V_txt, return_counts = True)

print("Alfabeto do texto:", alf)

#Obtenção da quantidade de símbolos do alfabeto
n_alf = len(alf)

#Conversão dos vetores em listas
alf = list(alf)
freq = list(freq)

#Delaração da tabela de códigos
cod = [""] * n_alf

#Cálculo da tabela de códigos pelo algoritmo de Huffman
huffman(alf, alf, freq, cod)

print("\nTabela de códigos (Huffman):", cod)

#Codificação do texto de acordo com a tabela de códigos obtida
final = code(string, alf, cod)

print("\nTexto codificado:", final)

#Cálculo da quantidade de bits necessária para representar a informação original sem compressão
bits_o = np.ceil(np.log2(n_alf)) * n_string

#Cálculo da quantidade de bits da representação da informação após compressão
bits_f = len(final)

#Cálculo da taxa de compressão
print("\nTaxa de compressão:", bits_o / bits_f)

#Cálculo da quantidade média de bits por símbolo após a compressão
print("\nQuantidade média de bits/caractere:", bits_f / n_string)

#Cálculo da entropia do texto original
print("\nEntropia do texto:", entropy(V_txt))
```
