```
#Função de ordenação de um vetor "V" em ordem decrescente de frequência
def ord(V):

  #Obtenção do vetor alfabeto e frequência referente a "V"
  alf, freq = np.unique(V, return_counts = True)

  #Listas que serão armazenados os resultados referentes a "V" em ordem decrescente
  alf_ord = []
  freq_ord = []

  for i in range(len(alf)):

    #Índice do elemento de máxima frequência
    max = np.argmax(freq)

    #Obtenção do elemento do alfabeto de maior frequência para ser inserido no vetor de alfabeto ordenado
    alf_ord.append(alf[max])
    
    #Obtenção da maior frequência para ser inserido no vetor de frequência ordenada
    freq_ord.append(freq[max])

    #Eliminação do elemento já analisado
    freq[max] = -1

  return alf_ord, freq_ord
```

```
#Função para realização da operação de diferença entre um vetor "V1" e um vetor "V2"
def setDiff(V1, V2):

  #Lista de armazenamento da operação de diferença
  V3 = []

  for i in range(len(V1)):

    #Cálculo da diferença entre "V1" e "V2"
    if V1[i] not in V2:
    
      V3.append(V1[i])
  
  return V3
```

```
#Função para realizar a soma cumulativa de uma lista de probabilidades "V"
def cumulative(p):

  #Lista para armazenamento da soma acumulada
  acum = []
  acum.append(0.0)

  #Obtenção da soma dos elementos do vetor
  sum = np.sum(p)

  #Cálculo da soma acumulada
  for i in range(len(p)):

    #Normalização do i-ésimo elemento para o cálculo da soma acumulada
    p[i] /= sum

    acum.append(acum[i] + p[i])

  return np.array(acum)
```

```
#Função para o cálculo da tabela de códigos de Shannon-Fano para um conjunto de dados "string", dado seu alfabeto "alf" e o vetor de códigos "cod" 
def shannonFano(string, alf, cod):

  #Conjunto para armazenamento do alfabeto esquerdo e direito (o alfabeto não contém símbolos repetidos)
  alf_L = []
  alf_R = []

  #Conjunto para armazenamento da string esquerda e direita (a string pode conter símbolos repetidos)
  string_R = []
  string_L = []

  #Obtenção do alfabeto e suas respectivas frequências em ordem decrescente
  string_ord, freq_ord = ord(string)

  #Obtenção da função de probabilidade acumulada do vetor de frequências
  prob = cumulative(freq_ord)

  #Índice de separação da string em parte direita e parte esquerda, de acordo com a função de probabilidade acumulada
  #O índice é calculado de forma a separar a string de entrada em vetores aproximadamente iguais em termos de probabilidades
  #Isto é, a separação ocorre de maneira a dividir o vetor de probabilidade acumulada a partir do termo mais próximo de 0.5
  ind = np.abs(prob - 0.5).argmin()

  for i in range(ind):

    #Obtenção do alfabeto esquerdo
    alf_L.append(string_ord[i])

  for i in range(ind, len(string_ord)):

    #Obtenção do alfabeto direito
    alf_R.append(string_ord[i])

  #Obtenção da string esquerda
  string_L = setDiff(string, alf_R)
  
  #Obtenção da string direita
  string_R = setDiff(string, alf_L)

  #Obtenção do tamanho do alfabeto esquerdo
  n_alf_L = len(alf_L)
  
  #Obtenção do tamanho do alfabeto direito
  n_alf_R = len(alf_R)

  #Analisa o alfabeto esquerdo, se houver elementos no mesmo
  if n_alf_L != 0:

    #Para cada símbolo do alfabeto esquerdo, adicionar o número 0 no código do mesmo
    for i in range(n_alf_L):

      cod[alf.index(alf_L[i])] += '0'
    
    #Se o alfabeto esquerdo tiver mais de um símbolo, realizar o mesmo processo para esse sub-conjunto
    if n_alf_L > 1:
    
      shannonFano(string_L, alf, cod)

  #Analisa o alfabeto direito, se houver elementos no mesmo
  if n_alf_R != 0:

    #Para cada símbolo do alfabeto direito, adicionar o número 1 no código do mesmo
    for i in range(n_alf_R):

      cod[alf.index(alf_R[i])] += '1'
    
    #Se o alfabeto direito tiver mais de um símbolo, realizar o mesmo processo para esse sub-conjunto
    if n_alf_R > 1:
    
      shannonFano(string_R, alf, cod)
```

```
#Codificação de um conjunto de dados "string", dado sua lista de códigos "cod" e seu alfabeto "alf"
def code(string, alf, cod):

  #String para armazenamento do resultado
  cod_string = ""

  for i in range(len(string)):

    #Obtenção do elemento corrente da string
    elem = string[i]
    
    #Obtenção da posição do elemento corrente no alfabeto
    pos = alf.index(elem)

    #Obtenção do código correspondente ao elemento corrente e junção do mesmo à codificação final
    cod_string += cod[pos]

  return cod_string
```

```
#Função que realiza o cálculo da entropia de um conjunto de dados armazenados em um vetor "V"
def entropy(V):

  #Obtenção da frequência de ocorrência de cada símbolo do alfabeto
  _, freq = np.unique(V, return_counts = True)

  #Obtenção do vetor de probabilidades, normalizando o vetor de frequências
  p = freq / len(V)

  #Variável para armazenamento da entropia
  etp = 0.0

  #Laço que analisa cada dado do vetor de probabilidades
  for i in range(len(p)):

    etp = etp - p[i] * np.log2(p[i])
  
  return etp
```

Utilizando o algoritmo de Shannon-Fano e as funções auxiliares definidas acima, 3 arquivos de texto distintos serão codificados.
Para medir a eficiência dessa técnica, será calculada a taxa de compressão dos mesmos e será realizada uma comparação da quantidade média de bits do resultado com a entropia da informação original.

```
#Texto I

#Carregamento do texto
txt = open(root_path + 'brandenburg.txt', 'r')

#Obtenção do conteúdo do arquivo
string = txt.read()
txt.close()

print("Texto original:", string)

#Obtenção do tamanho da string
n_string = len(string)

#Conversão da string em lista
V_txt = list(string)

#Obtenção do alfabeto e das frequências da string
alf, freq = np.unique(V_txt, return_counts = True)

print("Alfabeto do texto:", alf)

#Obtenção da quantidade de símbolos do alfabeto
n_alf = len(alf)

#Conversão dos vetores em listas
alf = list(alf)
freq = list(freq)

#Delaração da tabela de códigos
cod = [""] * n_alf

#Cálculo da tabela de códigos pelo algoritmo de Shannon-Fano
shannonFano(V_txt, alf, cod)

print("\nTabela de códigos (Shannon-Fano):", cod)

#Codificação do texto de acordo com a tabela de códigos obtida
final = code(string, alf, cod)

print("\nTexto codificado:", final)

#Cálculo da quantidade de bits necessária para representar a informação original sem compressão
bits_o = np.ceil(np.log2(n_alf)) * n_string

#Cálculo da quantidade de bits da representação da informação após compressão
bits_f = len(final)

#Cálculo da taxa de compressão
print("\nTaxa de compressão:", bits_o / bits_f)

#Cálculo da quantidade média de bits por símbolo após a compressão
print("\nQuantidade média de bits/caractere:", bits_f / n_string)

#Cálculo da entropia do texto original
print("\nEntropia do texto:", entropy(V_txt))
```
