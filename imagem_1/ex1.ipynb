```
#Função que realizará a separação dos 8 bits de cada pixel para a construção do bit plane.
#Cada matriz do bit plane é formada apenas por 1 bit, onde a n-ésima matriz é formada pelos n-ésimos bits de cada pixel.
#O bit plane resultante terá 8 matrizes, pois cada pixel é formada por 8 bits (nível de cinza variando de 0 a 255).
def getBitPlane(im):

  #Carregamento dos pixels da imagem
  pixel = im.load()

  #Obtenção do tamanho da imagem
  n_rows, n_columns = im.size

  plane = np.empty(im.size)

  #Número de bits da representação de imagem
  k = 8

  #Vetor de matrizes de 8 posições, onde as 8 matrizes resultantes serão carregadas
  bit_plane = np.array([plane] * k)

  #Laço que analisará cada pixel da imagem, realizando a extração de cada bit de cada pixel
  for i in range(n_rows):

    for j in range(n_columns):

      for n in range(k):
        
        #Operação que retornará o n-ésimo bit de cada pixel e armazenará na posição (i, j) da matriz n.
        #O operador ">>" realiza um deslocamento para a direita da representação do algarismo em binário.
        #Dessa forma, o valor do pixel em binário será deslocado para a direita em n vezes.
        #Após, é realizada uma operação AND entre esse resultado e o número 1 (00000001 em binário).
        #Logo, o resultado será somente o último bit do pixel, que foi deslocado n posições.
        #Portanto, o resultado será o n-ésimo bit do algarismo que representa a tonalidade do bit, contado do mais para o menos significativo.
        bit_plane[n, i, j] = (pixel[i, j] >> n) & 1

  return bit_plane
```

```
#Programa principal que executará a função "getBitPlane" e retornará 8 matrizes, representantes do bit plane da imagem.
from google.colab import drive
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

drive.mount('/content/gdrive')

root_path = '/content/gdrive/My Drive/Colab Notebooks/dados/'

#Carregamento da imagem
im = Image.open(root_path + 'lena.tiff')

#Execução da função 
bit_plane = getBitPlane(im)
```
