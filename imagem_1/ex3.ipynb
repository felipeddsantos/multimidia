```
#Função que normaliza uma matriz, formando matrizes resultantes de várias quantizações.
#Esse procedimento é equivalente a remover 1 bit de cada pixel, reduzindo seu alcance de representação e a variação de tons de cinza.
#Como resultado, esse processo gera 8 matrizes, cada uma com um bit a menos de representação em relação a anterior.
#Dessa forma, a primeira matriz tem 8 bits (0 : 255), a segunda tem 7 bits (0 : 127), a terceira tem 6 (0 : 63), até a última matriz, que tem apenas 1 bit (imagem preto e branco).
import math

def norm(pixel, n_rows, n_columns):

  plane = np.empty(im.size)

  #Número de bits da representação de imagem
  k = 8

  #Vetor de matrizes de 8 posições, onde as 8 matrizes resultantes serão carregadas
  quantization = np.array([plane] * k)

  #Laço que analisa todos os pixels e realiza as normalizações devidas, reduzindo a quantidade de bits e armazenando na matriz resultante
  for i in range(n_rows):

    for j in range(n_columns):

      for n in range(k):

        #Operação principal: todos os pixels são divididos por 2^n, reduzindo o intervalo de representação para (0 : 2^n - 1). 
        #Assim, para cada incremento de "k", a representação binária diminui pela metade.
        #O valores obtidos são arredondados para baixo, pois essa representação suporta somente números inteiros 
        quantization[n, i, j] = math.ceil(pixel[i, j] / (2 ** n))

  return quantization
```

```
#Programa para cálculo de quantizações de uma imagem, desde 8 bits (imagem original) até 1 bit (imagem binária)
from google.colab import drive
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

drive.mount('/content/gdrive')

root_path = '/content/gdrive/My Drive/Colab Notebooks/dados/'

#Obtenção da imagem
im = Image.open(root_path + 'ctskull.tif')

#Obtenção da matriz de pixels da imagem
pixel = im.load()

#Obtenção das dimensões da imagem
n_rows, n_columns = im.size

#Realização do processo de normalização, reduzindo a quantidade de bits da imagem
final = norm(pixel, n_rows, n_columns)
```
