```
#Função que realiza o cálculo do histograma normalizado de um conjunto de dados armazenados em uma matriz "M", dado o nível de quantização "quant"
#Para uma imagem em níveis de cinza de 8 bits, "quant" é igual a 256
def getHistogram(M, quant):

  #Obtenção das dimensões da imagem
  n_rows, n_columns = M.shape

  #Transformação dos elementos da matriz em inteiro
  M = M.astype('uint8')

  #Vetor que será armazenado o histograma, com o tamanho igual a quantidade de níveis disponíveis de acordo com a quantização
  #Nesse caso, a quatização é de 8 bits, ou seja, 2^8 = 256 níveis distintos
  histogram = np.zeros(quant)
    
  #Laço que analisa cada pixel do vetor e realiza a contagem de cada valor e determina a quantidade de cada nível na imagem
  for i in range(n_rows):

    for j in range(n_columns):

      histogram[M[i, j]] += 1
  
  return histogram
```

```
#Função que realiza a soma cumulativa de um vetor de probabilidades "V", necessária para o cálculo da distribuição de probabilidade acumuldada
def cumulative(V):

  #A lista "sum" armazena a soma cumulativa do vetor "V", que armazena as probabilidades dos níveis pertencentes a imagem 
  sum = []
  sum.append(0.0)

  #Ao fim do laço, que percorre todo o vetor de probabilidades, a função retorna uma lista de probabilidade acumulada
  for i in range(len(V)):

    sum.append(sum[i] + V[i])

  return sum
```

No código a seguir, será selecionado um grupo de 3 imagens em nível de cinza, onde cada uma passará pelo seguinte processo:

Obtenção do histograma normalizado;

Obtenção do histograma acumulado.

```
#Imagem 01

#Declaração das bibliotecas necessárias, definição do diretório raíz e montagem do diretório
from google.colab import drive
import numpy as np
from PIL import Image
from numpy import asarray
import matplotlib.pyplot as plt

drive.mount('/content/gdrive')

root_path = '/content/gdrive/My Drive/Colab Notebooks/dados/threshold/'

#Carregamento da imagem
im = Image.open(root_path + 'lena.tiff')

#Mostrar a imagem
plt.imshow(im, 'gray', origin = 'upper')
plt.show()
```

```
#Imagem 01

#Transformação da imagem em array
pixel = asarray(im)

#Obtenção das dimensões da imagem
n_rows, n_columns = pixel.shape

#Obtenção da quantidade de pixels da imagem
n = n_rows * n_columns

#Obtenção do histograma
hist = getHistogram(pixel, 256)

#Normalização do histograma
hist = hist / n

#Plotagem do histograma normalizado
plt.plot(hist)
```

```
#Imagem 01

#Obtenção do histograma acumulado
sum = cumulative(hist)

#Plotagem do histograma acumulado
plt.plot(sum)
```
