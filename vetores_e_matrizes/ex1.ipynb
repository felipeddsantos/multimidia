```
#a)
A = [[1, 4, 5, 12], [-5, 8, 9, 0], [-6, 7, 11, 19]]
print(A)
```

```
#b)
B = [[16, 8, 2, 4],
     [5, 7, 8, 11],
     [20, 30, 40, 50]]
print(B)
```

```
import numpy as np

#c)
C = np.array([[1, 2, 3], [3, 4, 5]])
print(C)
```

```
#d)
D = [[1, 2, 3, 4], [5, 6, 7, 8], [9, 8, 8, 9], [8, 7, 7, 8], [4, 5, 9, 8]]
print(D)
```
