```
#b)
import numpy as np

#Construção de matriz 3x4 formada por inteiros aleatórios
x2 = np.random.randint(10, size=(3, 4))
print(x2)
```

```
#Todos os elementos da coluna 0 da matriz x2
print(x2[:, 0])
```

```
#Todos os elementos da linha 0 da matriz x2
print(x2[0, :])
```

```
#Todos os elementos da linha 0 até a linha 2 e da coluna 2 até a coluna 3 da matriz x2
print(x2[0:2, 2:3])
```

```
#Fazer todos os elementos da coluna 0 iguais a zero
A4 = x2
A4[:, 0] = 0
print(A4)
```

```
#O elemento da última linha e última coluna da matriz x2
A5 = x2[-1, -1]
print(A5)
```

```
#Todos os elementos da última coluna da matriz x2
A6 = x2[:, -1]
print(A6)
```

```
#O elemento da última linha e primeira coluna antes da última da matriz x2
A7 = x2[-1, -2]
print(A7)
```

```
#Matriz transposta de x2
A8 = np.transpose(x2)
print(A8)
```

```
#c)

#I) Vetor com todos os valores de 1 a 20
v1 = np.arange(1, 21)
print(v1)
```

```
#II) Vetor com todos os valores pares de 1 a 20
v2 = np.arange(2, 21, 2)
print(v2)
```

```
#III) Vetor com todos os valores ímpares de 1 a 20
v3 = np.arange(1, 20, 2)
print(v3)
```

```
#IV) Vetores com valores de 0 a 1, intervalados de 0.1
v4 = np.arange(0, 1.1, 0.1)
print(v4)
```

No item 'c', foi utilizada a função "arange" para a construção dos vetores. A função "arange" constroi arrays com elementos dentro de um dado intervalo semi aberto, espaçados por um valor. Essa função possui a seguinte sintaxe: "arange(start, end, step)", onde:

"start" é o número inicial do intervalo (incluso no intervalo);

"end" é o número final do intervalo (não incluso no intervalo, portanto, para incluí-lo, deve fazer "end = final + 1");

"step" é o intervalo entre os números gerados.
