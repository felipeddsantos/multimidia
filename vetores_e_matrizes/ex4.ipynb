```
#b)

#Construção da matriz proposta no exercício matriz proposta no exercício
array_1 = np.arange(2, 41, 2)
M = np.array([array_1] * 15).astype(np.uint8)
print(M)

#Conversão da matriz M, construída anteriormente, em escala de cinza
plt.imshow(M,'gray')
plt.show()
img1 = Image.fromarray(M)
img1.save(root_path + 'dados/teste2.png')
```

Nessa escala de cinza, números menores correspondem a faixas mais escuras (próximas do preto). Consequentemente, números maiores correspondem a faixas mais claras (próximas do branco). Observa-se que o primeiro resultado (item a) foi uma imagem 10x20 (mesma ordem da matriz "array"), onde cada pixel corresponde a um elemento da matriz. Como o vetor "array" foi construído com números pseudoaleatórios, a imagem resultante foi uma série de pixels de diferentes tonalidades distribuídos aleatoriamente. Similarmente, o segundo resultado (item b) foi uma imagem 15x20 (mesma ordem da matriz 'M'), onde cada pixel corresponde a um elemento da matriz. Dada a estrutura da matriz 'M' (valores crescentes ao longo das colunas), a imagem resultante é uma série de faixas de diferentes tons de cinza, começando do valor 2 (mais escuro), até o valor 40 (mais claro).
