```
import numpy as np
from random import randint
from random import random
from random import seed

#a) Matriz de zeros de ordem 4
print(np.zeros((4,4)))
```

```
#b) Matriz de uns de ordem 4
print(np.ones((4,4)))
```

```
#c) Matriz de booleanos "True" de ordem 4
print(np.full((4,4), True))
```

```
#d) Gera um número pseudoaleatório entre 0 e 1, distribuído uniformemente
print(random())
```

```
#e) Matriz de números pseudoaleatórios de ordem 4, distribuídos normalmente
print(np.random.randn(4,4))
```

```
#f)

#I) 

#Cria um vetor de dimensão 5, com todos os elementos igual a zero, de tipo inteiro
Z1 = np.zeros((5, ), dtype = int)
print(Z1)
```

```
#II) 

#Cria um vetor de dimensão 5, com todos os elementos igual a um, de tipo float
Z2 = np.ones(5)
print(Z2)
```

```
#III) 

#Gera um número pseudoaleatório inteiro, entre 2 e 5 (incluindo 2 e 5)
value = randint(2, 5)
print(value)
```

```
#IV) Altera a semente do gerador de números pseudoaleatórios para 1. Dessa forma, se a semente sempre for a mesma, o gerador irá selecionar sempre o mesmo número
#Para que o gerador selecione números diferentes, é necessário alterar o valor da semente
#Assim, o comando "print(random())" irá mostrar sempre o mesmo número, como observado no item "V"

seed(1)
```

```
#V) Gera um número pseudoaleatório uniformemente distribuído 
#Devido ao fato de a semente estar em 1 pelo comando "seed(1)" do item "IV", esse número gerado será o mesmo

print(random())
```

```
#VI) Gera um vetor de dimensão 5 de números pseudoaleatórios normalmente distribuídos

np.random.randn(5)
```
