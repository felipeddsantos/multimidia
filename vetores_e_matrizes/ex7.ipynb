```
#Abrindo uma imagem com SciPy e cortando a imagem em 50 pixels de cada lado
from scipy import misc
import imageio
import matplotlib.pyplot as plt
from PIL import Image
from PIL import ImageOps

root_path = '/content/gdrive/My Drive/Colab Notebooks/'

#Abrindo a imagem
img_3 = Image.open(root_path + "dados/rose_gray.tif")

#Estabelecendo corte de 50 pixels de cada lado
borda = (50, 50, 50, 50)

#Cortando a imagem
img_4 = ImageOps.crop(img_3, borda)

#Salvando a imagem resultante em um arquivo "test3.tif"
img_4.save(root_path + 'dados/teste3.tif')

#Abrindo a imagem resultante para visualizar o resultado
img_4 = imageio.imread(root_path + 'dados/teste3.tif')
plt.imshow(img_4, cmap=plt.cm.gray)
plt.show()
```
