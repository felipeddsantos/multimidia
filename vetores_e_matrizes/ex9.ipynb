```
import matplotlib.pyplot as plt
import numpy as np

#a) Definição do intervalo (-2pi, 2pi)
x = np.arange(-2*np.pi, 2*np.pi, 0.1)

#Definição da função (sen(x) + 2)
y = np.sin(x) + 2

#Plotagem do gráfico
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```
#b) Definição do intervalo (0, 100)
x = np.arange(0, 100, 0.1)

#Definição da função (2x + 3)
y = 2 * x + 3

#Plotagem do gráfico
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```
#c) Definição do intervalo (-100, 100)
x = np.arange(-100, 100, 0.1)

#Definição da função (x^2 +4x - 17)
y = x * x + 4 * x - 17

#Plotagem do gráfico
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
