```
#Manipulação do primeiro quadro do vídeo

#Declaração das bibliotecas necessárias, definição do diretório raíz e montagem do diretório
from google.colab import drive
import cv2
from cv2 import imwrite
import matplotlib.pyplot as plt

drive.mount('/content/gdrive')

root_path = '/content/gdrive/My Drive/Colab Notebooks/dados/'

#Carregamento do vídeo
vid = cv2.VideoCapture(root_path + 'bird.avi')

#Captura do primeiro quadro
#Cada vez que o método "read" é invocado, o próximo quadro do vídeo é carregado para a variável "frame"
#Nesse caso, uma chamada desse método carrega somente o primeiro quadro
ret, frame = vid.read() 

vid.release()  
 
#Plotagem do primeiro quadro
plt.imshow(frame, 'gray', origin = 'upper')
plt.show()
```

```
#Obtenção da estrutura do primeiro quadro pelo método "shape", que retorna as dimensões da imagem e a quantidade de canais
#Nesse caso, o quadro possui dimensões de 240x256 e 3 canais
print(frame.shape)
```

```
#Conversão do quadro obtido anteriormente para o sistema de cores HSV
frame_hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

#Salvando a imagem resultante
imwrite(root_path + "frame_1.jpg", frame_hsv);

#Mostrar a imagem resultante
plt.imshow(frame_hsv, 'gray', origin = 'upper')
plt.show()
```
