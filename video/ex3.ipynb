```
#Reprodução do vídeo

#Declaração das bibliotecas necessárias, definição do diretório raíz e montagem do diretório
import cv2
from google.colab import drive

drive.mount('/content/gdrive')

root_path = '/content/gdrive/My Drive/Colab Notebooks/dados/'

#Carregamento do vídeo
vid = cv2.VideoCapture(root_path + 'bird.avi')

#Obtenção da quantidade de quadros do vídeo
n_frames = int(vid.get(cv2.CAP_PROP_FRAME_COUNT))

#Obtenção da taxa de quadros do vídeo
fps = int(vid.get(cv2.CAP_PROP_FPS))

#Tempo de pausa entre a exibição de um quadro e outro, em milissegundos
t = int(1000 / fps)

#Variável para contagem dos quadros exibidos
count = 0

#Laço que exibirá todos os quadros do vídeo
while count < n_frames: 

    #Obtenção do próximo quadro do vídeo
    ret, frame = vid.read() 
 
    #Exibição do quadro corrente
    cv2.imshow('bird.avi', frame) 

    #Pausa até a obtenção e exibição do próximo frame, dada em milissegundos
    #Controla a quantidade de frames que serão exibidos por segundo, de acordo com a variável "t" definida anteriormente
    cv2.waitKey(t)

    #Incremento da contagem de frames já exibidos   
    count += 1

vid.release() 
cv2.destroyAllWindows() 
```

```
#Reprodução de alguns quadros do vídeo

#Carregamento do vídeo
vid = cv2.VideoCapture(root_path + 'bird.avi')

#Somente 10 quadros do vídeo serão considerados
n_frames = 10

count = 0

#Laço que exibirá os 10 primeiros quadros do vídeo
while count < n_frames: 

    ret, frame = vid.read() 
 
    cv2.imshow('bird.avi', frame) 

    cv2.waitKey(t)

    count += 1

vid.release() 
cv2.destroyAllWindows() 
```

```
#Considerando uma taxa de quadros igual a "fps", 10 quadros levarão um tempo de "10 / fps" segundos para reprodução
print("Taxa de quadros por segundo:", fps)
print("Tempo para reprodução de 10 quadros (em segundos):", 10 / fps)
```

Verificamos que a exibição ocorre em menos de meio segundo. Logo, não somos capazes de perceber a reprodução desses 10 primeiros quadros, devido a alta taxa de quadros. Para tal, podemos alterar a taxa de exibição do vídeo para que possamos notar esses 10 quadros. Por exemplo, para uma taxa de quadros de 2 quadros por segundo, a reprodução dos 10 primeiros quadros terão duração de 5 segundos.

```
#Reprodução de alguns quadros do vídeo em uma taxa de quadros diferente

#Carregamento do vídeo
vid = cv2.VideoCapture(root_path + 'bird.avi')

#Alteração da taxa de quadros do vídeo
fps = 2

#Tempo de pausa entre a exibição de um quadro e outro, dado o novo valor de "fps"
t = int(1000 / fps)

n_frames = 10

count = 0

#Laço que exibirá os 10 primeiros quadros do vídeo, a uma taxa de 2 quadros por segundo
while count < n_frames: 

    ret, frame = vid.read() 
 
    cv2.imshow('bird.avi', frame) 

    cv2.waitKey(t)

    count += 1

vid.release() 
cv2.destroyAllWindows() 
```
